### Libraries

```python
# Libraries
import textworld
from textworld import GameMaker
```

### Easy Difficulty

```python
# Create Game Enviroment
GM_Easy = GameMaker()

# Add rooms to game enviroment
bedroom = GM_Easy.new_room("Bedroom") # For some reason, "Prison" doesn't work
kitchen = GM_Easy.new_room("Kitchen")
living_room = GM_Easy.new_room('LivingRoom')
corridor = GM_Easy.connect(bedroom.east, living_room.west) # Connet rooms
corridor2 = GM_Easy.connect(living_room.north, kitchen.south)

# Set player
GM_Easy.set_player(bedroom)

# Supporters
table = GM_Easy.new(type='s', name="table")
bed = GM_Easy.new(type='s', name='bed')
counter = GM_Easy.new(type='s', name='counter')
# Things
phone = GM_Easy.new(type='o', name='phone')
banana = GM_Easy.new(type='o', name='banana')
blanket = GM_Easy.new(type='o', name='blanket')
# Add supporters in room
bedroom.add(bed)
living_room.add(table)
# Add things on supporters
bed.add(blanket)

# Add doors
bedroom_door = GM_Easy.new_door(corridor, name="door")
GM_Easy.add_fact("closed", bedroom_door)

GM_Easy.render() # Render
```

### Hard Difficulty

```python
# Create Game Enviroment
GM_Hard = GameMaker()

# Add rooms to game enviroment
bedroom = GM_Hard.new_room("Bedroom") # For some reason, "Prison" doesn't work
kitchen = GM_Hard.new_room("Kitchen")
living_room = GM_Hard.new_room('LivingRoom')
toilet = GM_Hard.new_room('Toilet')
corridor = GM_Hard.connect(bedroom.east, living_room.west) # Connet rooms
corridor2 = GM_Hard.connect(living_room.north, kitchen.south)
corrider3 = GM_Hard.connect(living_room.south, toilet.north)

# Set player
GM_Hard.set_player(bedroom)

# Supporters
table = GM_Hard.new(type='s', name="table")
bed = GM_Hard.new(type='s', name='bed')
counter = GM_Hard.new(type='s', name='counter')
# Things
phone = GM_Hard.new(type='o', name='phone')
banana = GM_Hard.new(type='f', name='banana')
GM_Hard.add_fact("edible", banana)
blanket = GM_Hard.new(type='o', name='blanket')
soap = GM_Hard.new(type='o', name='soap')
toilet_paper = GM_Hard.new(type='o', name='toilet paper')
# Add supporters in room
bedroom.add(bed)
living_room.add(table)
kitchen.add(counter)
# Add things on supporters
bed.add(blanket)
table.add(phone)
counter.add(banana)
# Add things on the floor
toilet.add(toilet_paper)
toilet.add(soap)

# Add doors
bedroom_door = GM_Hard.new_door(corridor, name="door")
GM_Hard.add_fact("closed", bedroom_door)

GM_Hard.render() # Render
```
