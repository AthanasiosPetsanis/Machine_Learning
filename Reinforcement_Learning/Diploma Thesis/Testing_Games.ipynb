```python
# Libraries
import gym
import textworld.gym
from textworld.generator import make_game, compile_game
from textworld import EnvInfos

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Register a text-based game as a new Gym's environment.


# --- Make random game ---

# options = textworld.GameOptions()
# options.path = "./Random_Games/"
# options.seeds = 432
# game = make_game(options)
# game_file = compile_game(game, options)
# print(game_file)
# -----------------------------


# --- Make random game with quick command ---

# THIS DOES NOT OVERWRITE GAMES WITH EXISTING NAMES!
# !tw-make custom --world-size 2 --nb-objects 5 --theme house --quest-length 4 --quest-breadth 2 --entity-numbering --output "./Created_Games/MDP_Game.ulx" --seed 5 --format ulx

# -------------------------------------

# Load Game
max_steps = 50
# game_file = "./games/rewardsDense_goalBrief.ulx"
game_file = "./Created_Games/Realistic_Game_Hard.ulx"

request_infos = EnvInfos(inventory=True, admissible_commands=True, policy_commands=True, entities=True, won=True, lost=True,
                         description=True, location=True, objective=True, score=True, moves=True)

env_id = textworld.gym.register_games([game_file], request_infos)

env = gym.make(env_id)  # Start the environment.

obs, infos = env.reset()# Start new episode.

print(obs)

score, moves, done = 0, 0, False
```

```python
# infos["location"] always returns None for some reason
# So we need to get it manually from description

def find_location(desc):
    location = ""
    counter = 0

    for letter in desc:

        if counter==1 and letter!=" ":
            location += letter

        if letter==" ":
            counter += 1
    
        if counter==2:
            return location
    
# Also we don't need the "inventory" and "look" commands
def cmd_remover(ad_cmds):
    ad_cmds.remove("inventory")
    ad_cmds.remove("look")
```

```python
print('Actions towards goal: ', end="")
for commands in infos['policy_commands']:
    print(commands, end=" > ")
```

```python
# Play game
print(infos['policy_commands'])
while not done:
    cmd_remover(infos["admissible_commands"])
    infos["location"] = find_location(infos["description"])
    print("Location: {}, Score: {}".format(infos["location"], score))
    print("Entities: \n{}".format(infos["entities"]))
    print("Inventory: {}".format(infos["inventory"]))
    print("Possible actions: \n{}".format(infos["admissible_commands"], end="\n"))
    command = input("> ")
    obs, score, done, infos = env.step(command)
    print("\nDONE: {}\n".format(done))
    print("\nWON: {}\n".format(infos["won"]))
    env.render()

env.close()
print("moves: {}; score: {}".format(infos["moves"], score))
```
