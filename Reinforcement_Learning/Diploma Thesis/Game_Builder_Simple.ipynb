### Easy Difficulty

```python
# Create Game Enviroment
GM_Easy = GameMaker()

# Add rooms to game enviroment
bedroom = GM_Easy.new_room("Bedroom") # For some reason, "Prison" doesn't work
kitchen = GM_Easy.new_room("Kitchen")
corridor = GM_Easy.connect(bedroom.east, kitchen.west) # Connet rooms

# Set player
GM_Easy.set_player(bedroom)

# Add things in one room
table = GM_Easy.new(type='s', name="table")
glasses = GM_Easy.new(type='o', name='glasses')
bedroom.add(table)
table.add(glasses)

GM_Easy.render() # Render
```

### Medium Difficulty

```python
# Create Game Enviroment
GM_Medium = GameMaker()

# Add rooms to game enviroment
bedroom = GM_Medium.new_room("Bedroom") # For some reason, "Prison" doesn't work
kitchen = GM_Medium.new_room("Kitchen")
corridor = GM_Medium.connect(bedroom.east, kitchen.west) # Connet rooms

# Set player
GM_Medium.set_player(bedroom)

# Add things in one room
table = GM_Medium.new(type='s', name="table")
glasses = GM_Medium.new(type='o', name='glasses')
bedroom.add(table)
table.add(glasses)

# Add things in second room
water = GM_Medium.new(type='o', name="bottle of water")
apple = GM_Medium.new(type='f', name="apple")
GM_Medium.add_fact("edible", apple)
table2 = GM_Medium.new(type='s', name="table")
kitchen.add(table2)
table2.add(apple, water)

GM_Medium.render() # Render
```

### Hard Difficulty

```python
# Create Game Enviroment
GM_Hard = GameMaker()

# Add rooms to game enviroment
bedroom = GM_Hard.new_room("Bedroom") # For some reason, "Prison" doesn't work
kitchen = GM_Hard.new_room("Kitchen")
corridor = GM_Hard.connect(bedroom.east, kitchen.west) # Connet rooms

# Set player
GM_Hard.set_player(bedroom)

# Add things in one room
table = GM_Hard.new(type='s', name="table")
glasses = GM_Hard.new(type='o', name='glasses')
bed = GM_Hard.new(type='s', name='bed')
key = GM_Hard.new(type='k', name="key")
bedroom.add(table)
table.add(key)
table.add(glasses)

# Add locked door that opens with specific key
door = GM_Hard.new_door(corridor, name="door")
GM_Hard.add_fact("locked", door)
GM_Hard.add_fact("match", key, door)

# Add things on player
money = GM_Hard.new(type='o', name='money')
GM_Hard.inventory.add(money)

# Add things in second room
water = GM_Hard.new(type='o', name="bottle of water")
apple = GM_Hard.new(type='f', name="apple")
GM_Hard.add_fact("edible", apple)
table2 = GM_Hard.new(type='s', name="table")
kitchen.add(table2)
table2.add(apple, water)

GM_Hard.render() # Render
```

### Record Goals

```python
# Create Goal
goal = []
goal.append(GM_Easy.record_quest())
goal.append(GM_Medium.record_quest())
goal.append(GM_Hard.record_quest())

for gl in goal:
    print(" > ".join(gl.commands))
    print("\n" + gl.desc)
```
